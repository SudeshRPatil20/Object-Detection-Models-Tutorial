```python
import os
```

```python
# hear we create a function that check for how many data does have in file
def get_file(file_path):
  if os.path.exists(file_path) and os.path.isdir(file_path):
    files = [f for f in os.listdir(file_path) if os.path.isfile(os.path.join(file_path, f))]
    num_file = len(files)
    print(f"Number of files in : {file_path}: {num_file}")
  else:
    print(f"Folder '{file_path} does not exis in directory.")
```

```python
from genericpath import isdir
def check_files(dir_path, structure=['images', 'labels']):
  folders = os.listdir(dir_path)
  for folder in folders:
    folder_path = os.path.join(dir_path, folder)
    if os.path.isdir(folder_path):
      sub_folder= os.listdir(folder_path)
      for final_folder in sub_folder:
        if final_folder in structure:
          get_file(os.path.join(folder_path, final_folder))
```

```python
import zipfile
import os

def zip_folder(folder_path, output_zip_path):
  try:
    with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
      for root, dirs, files in os.walk(folder_path):
        for file in files:
          file_path = os.path.join(root, file)
          arc_name = os.path.realpath(file_path, folder_path)
          zipf.write(file_path, arc_name)
    print(f"Folder '{folder_path}' zip sucessfully to '{output_zip_path}'")
  except FileNotFoundError:
    print(f"Error : File not Found {folder_path}")
  except Exception as e:
    print(f"An error occured: {e}")
```
